```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# load label CSV
df = pd.read_csv('train.csv')
df['path'] = 'train/train/' + df['id']

# get 0 and 1 label
has_cactus_df = df[df['has_cactus'] == 1].sample(5, random_state=42)
no_cactus_df = df[df['has_cactus'] == 0].sample(5, random_state=42)

# visualization
def show_images(df_subset, title):
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))
    fig.suptitle(title, fontsize=16)
    for i, (idx, row) in enumerate(df_subset.iterrows()):
        img = Image.open(row['path'])
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(f"Label: {row['has_cactus']}")
    plt.tight_layout()
    plt.show()

# show image with cactus
show_images(has_cactus_df, "Images with Cactus")

# show image without cactus
show_images(no_cactus_df, "Images without Cactus")
```
